#### Naming and Using Variables in Python

When naming a variable, ensure that its name is clear, readable, and descriptive of its purpose. Additionally, there are rules and guidelines you should follow when naming variables to avoid errors and prevent misleading or confusing names:

- A variable name can contain **letters, numbers, and underscores (`_`)**.

- A variable name must **start with a letter or an underscore**, not a number such as `1_var`.

- **Spaces are not allowed** in variable names. Use underscores to separate words.

- Do **not use Python keywords or built-in function names** (like `print`, `list`, or `type`) as variable names.

- Use **short but descriptive names** that explain what the variable stores.

- Avoid confusing characters such as lowercase **l** and uppercase **O**, because they can be confused with the numbers 1 and 2.

# If Statement
You must already experienced the conditional statement in MATLAB, so that you are familiar with its uses and purposes. In Python, the `if` statement serves the same purpose, it allows a program to make decisions and execute certain code only when a specified condition is true.  
The basic syntax of an `if` statement in Python is:  
`if condition:`  
&ensp; &ensp; &ensp;`statement`

- No parentheses needed around the condition.
- A colon (`:`) is essential at  the end of the `if` line.
- The alignment is important.
- There is no `end` keyword.

```python
x = 6
if x > 10: 
    print("x is greater than 10") 
elif x > 5: 
    print("x is greater than 5 but not greater than 10") 
else: 
    print("x is 5 or less")
```

```python
def calculate(a, b, operation):
    def add():
        return a + b
    
    def subtract():
        return a - b
    
    def multiply():
        return a * b
    
    def divide():
        return a / b
    
    # For the third argument we need.
    if operation == 'add':
       return add()
    elif  operation == 'subtract':
        return subtract()
    elif operation == 'multiply':
        return multiply()
    elif operation == 'divide':
        return divide()

calculate(4, 2, 'divide')
```

```python
def is_numeric(number):
    if type(number) == int or type(number) == float:
        return True
    return False

is_numeric('s')
```

# Loops
In Python, loops allow us to repeat a block of code multiple times. The two main types of loops are the `for` loop and the `while` loop.
The synatx of the loop in Python will be:

`while [condition]:`

&ensp; &ensp; &ensp; `[Body (instructions)]`

### - When to use each?

<div align="center">

| Situation | Use |
|----------|-----|
| You know the number of iterations | `for` loop |

| You loop until a condition changes | `while` loop |

```python
counter = 0
while counter < 5:
    print(counter)
    counter += 1
```

# Object-Oriented Programming

**Object-Oriented Programming (OOP)** is a programming paradigm that organizes code around **objects** and **classes** rather than functions and logic alone. It focuses on creating **reusable**, **modular**, and scalable code by modeling real-world entities. 

In OOP, data (`attributes`) and behavior (`methods`) are bundled together into classes, which act as blueprints for creating objects. These objects interact with each other to perform various tasks, similar to how real-world entities interact.

## Why we need OOP?

- Helps you **reuse code** using concepts like **inheritance** (creating new classes from existing ones).
- Makes code easier to understand by organizing it into **small, focused parts**.
- Simplifies fixing bugs and updating programs.
- Keeps data safe by controlling how it’s accessed.
- Allows flexible code where objects can behave in different ways.

Objects can access class attributes also from the method.

```python
class Cookie(object):
    defalut_scarf_color = 'green'
    def __init__(self, buttons, scarf = None):
        # Scarf is Optional
        self.buttons = buttons
        if scarf:
            self.scarf = scarf
        else: 
            # Access the class defalut attribute
            self.scarf = self.defalut_scarf_color
```

```python
cookie_1 = Cookie('black')
cookie_2 = Cookie('tale', 'Pink')
cookie_3 = Cookie(buttons= 'yellow', scarf= 'white')
```

```python
class Cookie(object):
    default_scarf_color = 'green'
    defalut_cookie_weight = 100

    def __init__(self, buttons, scarf=None, weight = None):
        self.buttons = buttons
        if scarf:
            self.scarf = scarf
        else:
            self.scarf = self.default_scarf_color
        if weight:
            self.weight = weight
        else: 
            self.weight = self.defalut_cookie_weight

    def chop_in_half(self):
        self.weight = self.weight / 2
        return "The cookie has been chopped in half!"
```

```python
cookie_1 = Cookie('black')
cookie_2 = Cookie('tale', 'Pink', 300)
cookie_3 = Cookie(buttons= 'yellow', scarf= 'white')
```
