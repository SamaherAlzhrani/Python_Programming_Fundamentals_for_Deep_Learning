# Vectorized Operations with NumPy

"Vectorization" is an important concept in numeric computing. It refers to the process of operating in entire arrays, in a concise and efficient way. Let's take a look with a couple of examples.

Numpy also refers to _vectorized operations_ as **"array broadcasting"**. 

```python
import numpy as np
```

```python
a = np.arange(0, 5)
a
```

```python
b = np.arange(5, 10)
b
```

A vectorized operation is any algebraic binary operation you perform on them:

```python
a + b # elementwise operation
```

In this case, the operation was performed "element-wise", ie: "vectorized". Let's compare it with a regular for-loop operation:

```python
res = []
for i in range(len(a)):
    res.append(a[i] + b[i])
np.array(res)
```

It's much easier to express the operation as `a + b`, instead of explicitly defining the for loop to perform the element-wise sum. This is the difference between _declarative_ and _imperative_ programming.

Finally, compare this to the usage of regular Python lists:

```python
l1 = list(range(5))
l1
```

```python
l2 = list(range(5, 10))
l2
```

```python
l1 + l2 # sum of lists is not elementwise summing it just adding new elements to new list of sum
```

Python lists don't operate "algebraically", the `+` operation is just concatenation.

Generally, array vectorization will always work with arrays of the same size and shape, but it can also work for arrays of different shapes. For example:

```python
A = np.array([
    [0, 0, 0],
    [1, 1, 1],
    [2, 2, 2],
    [3, 3, 3]
])
A
```

```python
A.shape
```

```python
b = np.array([[1, 2, 3]])
b
```

```python
b.shape
```

Even though the shapes are different, we can still perform a vectorized operation between `A` and `b`:

```python
A + b # it did elementwise summing for each row in A 
```

This is possible since:

> **The Broadcasting Rule**: Two arrays are compatible for broadcasting if for each _trailing dimension_ (i.e., starting from the end) the axis lengths match or if either of the lengths is 1. Broadcasting is then performed over the missing or length 1 dimensions.

Any operation between arrays not following _The Broadcasting Rule_, will fail with a "broadcast" related error. For example:

```python
A = np.arange(6).reshape(2, 3)
A
```

```python
b = np.array([5, 5])
b
```

```python
A + b # we can't perform, can't broadcast because the shapes are not compatiable 
```

```python
a = np.arange(0, 6)
a
```

```python
a + 10 # perform over the whole array, this notion fails for lists.
```

As you can see, the `+ 10` operation was "broadcasted" (or "distributed") among all the elements of the array. This is what we usually do with a list comprehension in pure Python:

```python
[x + 10 for x in a]
```

But this "broadcasting" behavior is default for Numpy. Here are a few other examples:

```python
a - 5
```

```python
a * 3
```

## Numpy Boolean Array

This kind of arrays help us to perform some sort of filteration of data that we would use to machine learning.

```python
a > 2 # it did check element by element, it doesn't work with the lists
```

```python
a == 0 
```

```python
c = np.arange(0,20,3)
c
```

```python
c % 2 == 0 # Modular Arithmetic
```



A common practice is to use the function `np.sum` to count how many elements in the array satisfy a given condition. For example, from our array `a`, how many elements are greater than 2 (`> 2`)? The answer is `3` (elements `3`, `4` and `5`).

```python
a > 2 # 0 is False and 1 is True, so we can perform sum function.
```

```python
np.sum(a > 2)
```


Finally, it's important to note that vectorized operations (algebraic and boolean) with one dimensional arrays will also work for multidimensional arrays:

```python
A = np.arange(1, 10).reshape(3, 3)
A
```

```python
A + 100
```

```python
A % 2
```

```python
A % 2 == 0
```
